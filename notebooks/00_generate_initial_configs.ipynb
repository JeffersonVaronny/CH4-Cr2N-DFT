# Nomenclature
### H{config}\_Rot{degrees}\_{Site}.xyz 

- {config}: the first number designates the number of H atoms pointing downward, the second designates those pointing upward.
    - H22: 2 down, 2 up (original).
    - H31: 3 down, 1 up.
    - H13: 1 down, 3 up.

- {degrees}: angle in degrees by which the molecule is rotated with respect to the z-axis.

- {site}: type of adsorption site.
    - ontop: on top of the atom.
    - bridge: at the bridge position.
    - hollow: above the atom of the second layer (hollow site).

# Lectura de archivos xyz


```python
from pathlib import Path
from ase.io import read, write
from ase.visualize import view
```

```python
# Detectar carpeta base del proyecto
try:
    # Si corre como script
    base_dir = Path(__file__).resolve().parent.parent
except:
    # Si corre en Jupyter Notebook
    base_dir = Path.cwd().resolve().parent

# Definir rutas
structures_dir = base_dir / "structures"

reference_dir = structures_dir / "reference"

output_dir = structures_dir / "generated"

output_dir.mkdir(exist_ok=True)

# Crear carpetas si no existen
output_dir.mkdir(parents=True, exist_ok=True)
reference_dir.mkdir(parents=True, exist_ok=True)

print("Base dir:", base_dir)
```

```python
# Leer estructuras de referencia
slab = read(reference_dir / "slab.xyz")
methane = read(reference_dir / "CH4.xyz")
```

```python
# calcular las posiciones
positions = slab.get_positions()

ontop = slab.get_positions()[28]

bridge = 0.5 * ( positions[28] + positions[45])
#bridge_N = 0.5 * ( positions[28] + positions[63])

hollow = (1/3) * ( positions[28] + positions[45] + positions[65])
#hollow_N = (1/3) * ( positions[28] + positions[63] + positions[29])

sites_lis = [ontop, bridge, hollow]
sites_lis_nom = ["ontop", "bridge", "hollow"]
```

# H22\_Rot{grados}\_{Site}.xyz

```python
methane_H22 = methane.copy()

structures_H22 = []
contH22=0
adsorption_heightH22 = 2 + 0.6293

for i in range(3):
    for j in range(3):
        config, rotation, site = "22", j * 45, sites_lis[i]
        
        mol = methane_H22.copy()
        mol.rotate(rotation, 'z', center='COM')
        
        carbon_atom_pos = mol.get_positions()[0]
        final_carbon_pos = site + [0, 0, adsorption_heightH22]
        translation_vector = final_carbon_pos - carbon_atom_pos
        mol.translate(translation_vector)
        final_structure = slab + mol
        final_structure.center(axis=2, vacuum=5.0)
        
        output_filename = f"H{config}_Rot{rotation}_{sites_lis_nom[i]}.xyz"
        
        output_path = output_dir / output_filename
        write(output_path, final_structure)
        num = contH22 + 1
        print(f"{num} : Successfully created '{output_filename}'.")
        #print(f"Methane (atom 0) placed at {final_carbon_pos}")
        #print(f"Above slab atom {target_atom_index} (at {target_atom_pos})")
        print("---------------------------------------")
        
        structures_H22.append(final_structure)
        contH22+= 1
#view(structures_H22)
```

```python
# verificar que la diferencia de altura entre la superficie y el atomo de H más bajo sea 2 angstrom
pos_H22 = structures_H22[0].get_positions()
print(pos_H22[75])
print(pos_H22[28])
print(f"Height difference H <-> slab: {pos_H22[75][2] - pos_H22[28][2]}")
```
