````python
configs = Path("configs")

slab_file = configs / "clean_slab" / "Cr2N_slab.xyz"

configs = {
    "H31_Rot30_hollow": configs / "H31_Rot30_hollow.xyz",
    "H31_Rot60_hollow": configs / "H31_Rot60_hollow.xyz",
    "H22_Rot90_ontop": configs / "H22_Rot90_ontop.xyz",
    "H22_Rot45_ontop": configs / "H22_Rot45_ontop.xyz",
    "H22_Rot0_ontop": configs / "H22_Rot0_ontop.xyz",
    "H13_Rot0_ontop": configs / "H13_Rot0_ontop.xyz",
}
````

````python
slab = read(slab_file)

symbols = np.array(slab.get_chemical_symbols())
positions = slab.get_positions()
#view(slab)
````

````python
cr_indices = np.where(symbols == "Cr")[0]

cr_z = positions[cr_indices,2]

# índices de los 12 Cr con mayor z
top12 = cr_indices[np.argsort(cr_z)[-12:]]

# ordenarlos de izquierda a derecha para que siempre aparezcan igual
top12 = top12[np.argsort(positions[top12,0])]

print(top12)
````

````python
reference = pd.DataFrame({
    "Atom": top12,
    "x": positions[top12,0],
    "y": positions[top12,1],
    "z": positions[top12,2],
})

reference
````

````python
results = {}
references = {}

for name, xyzfile in configs.items():
    name = name.replace("_", " ").replace("0", "0°").replace("5", "5°").replace("Rot", "")
    atoms = read(xyzfile)

    pos = atoms.get_positions().copy()
    
    # ===== eliminar la traslación global =====
    shift = np.mean(pos[top12] - positions[top12], axis=0)
    pos -= shift
    # =========================================

    references[name] = pd.DataFrame({
        "Atom": top12,
        "x": pos[top12,0],
        "y": pos[top12,1],
        "z": pos[top12,2],
    })

    dx = pos[top12,0] - positions[top12,0]
    dy = pos[top12,1] - positions[top12,1]
    dz = pos[top12,2] - positions[top12,2]

    dr = np.sqrt(dx**2 + dy**2 + dz**2)

    df = pd.DataFrame({
        "Atom": top12,
        "Δx": dx,
        "Δy": dy,
        "Δz": dz,
        "Δr": dr
    })

    results[name] = df

display(results["H31 30° hollow"])
````

````python
summary = []

for name, df in results.items():

    # guardar desplazamientos completos
    df.to_csv(results_dir/f"displacement_{name}.csv",
              index=False)

    summary.append({
        "Configuration": name,
        "Max Δr (Å)": df["Δr"].max(),
        "Mean Δr (Å)": df["Δr"].mean(),
        "RMS Δr (Å)": np.sqrt(np.mean(df["Δr"]**2)),
        "Max Δz (Å)": df["Δz"].abs().max(),
        "Mean Δz (Å)": df["Δz"].mean()
    })

summary = pd.DataFrame(summary)


#summary.to_csv(results_dir/"displacement_summary.csv",index=False)
display(summary)
````

````python
ch4_ref = read("configs\CH4\methane.xyz")
#view(ch4_ref)
````

````python
def ch_bonds(atoms):
    """
    Devuelve las cuatro distancias C-H ordenadas de menor a mayor.
    """

    symbols = atoms.get_chemical_symbols()

    C = [i for i,s in enumerate(symbols) if s=="C"][0]
    H = [i for i,s in enumerate(symbols) if s=="H"]

    d = []

    for h in H:
        d.append(atoms.get_distance(C,h,mic=True))

    return np.sort(d)
````

````python
ref_CH = ch_bonds(ch4_ref)
print(ref_CH)
````

````python
CH_results = {}

for name, xyz in configs.items():
    name = name.replace("_", " ").replace("0", "0°").replace("5", "5°").replace("Rot", "")
    atoms = read(xyz)

    bonds = ch_bonds(atoms)

    CH_results[name] = bonds - ref_CH
````

````python
df_CH = pd.DataFrame(
    CH_results,
    index=["1.C-H","2.C-H","3.C-H","4.C-H"]
).T

display(df_CH)
````
